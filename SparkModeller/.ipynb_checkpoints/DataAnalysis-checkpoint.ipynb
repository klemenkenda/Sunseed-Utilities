```python
# importing
import datetime
import model
import numpy as np
import pandas as pd
```

```python
# load model data
infile = open('dataset-tr2-597.txt', 'rb')
dataset = np.load(infile)
infile.close()
```

```python
# creating attributes and targets
X = dataset[:,2:]
y = dataset[:, 1]
```

```python
# importing relevant libraries for regression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import linear_model
```

```python
# splitting dataset
n = len(X)
b_train = round(n * 0.75)
b_new = -25
X_train, X_test, X_new = X[0:b_train], X[b_train:b_new], X[b_new:]
y_train, y_test = y[0:b_train], y[b_train:b_new]
```

```python
gbr = GradientBoostingRegressor(n_estimators=100, 
                                learning_rate=0.3, max_depth=15, random_state=0, 
                                loss='ls').fit(X_train, y_train)
rfr = RandomForestRegressor(n_estimators = 90).fit(X_train, y_train)
lr = linear_model.LinearRegression().fit(X_train, y_train)
```

```python
mean_squared_error(y_test, gbr.predict(X_test))    
```

```python
mean_squared_error(y_test, lr.predict(X_test))
```

```python
mean_squared_error(y_test, rfr.predict(X_test))
```

```python
gbr_test = gbr.predict(X_test)
lr_test = lr.predict(X_test)
rfr_test = rfr.predict(X_test)
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt
```

```python
plt.plot(y_test)
plt.plot(gbr_test)
plt.plot(lr_test)
plt.plot(rfr_test)
```

## Multiple models (24)

```python
# create 24 datasets
l = len(dataset[0])
ds = np.empty(24, dtype=object);

for i in range(24):
    ds[i] = np.empty(l, dtype=object)
    # ds.append(np.empty(l, dtype=object))    

for i in range(len(dataset)):
    subset = i % 24
    if ds[subset][0] == None:
        ds[subset] = [dataset[i]]
    ds[subset] = np.append(ds[subset], [dataset[i]], axis=0)

# creating attributes and targets
X = np.empty(24, dtype=object)
y = np.empty(24, dtype=object)
rfr = np.empty(24, dtype=object)
result = np.empty(24, dtype=object)
for i in range(24):
    X[i] = ds[i][:,2:]
    y[i] = ds[i][:, 1]

    # splitting dataset
    n = len(X[i])
    b_train = round(n * 0.75)
    b_new = -1
    X_train, X_test, X_new = X[i][0:b_train], X[i][b_train:b_new], X[i][b_new:]
    y_train, y_test = y[i][0:b_train], y[i][b_train:b_new]
    
    # learn 24 models
    rfr[i] = RandomForestRegressor(n_estimators = 190).fit(X_train, y_train)
    
    # predict with 24 models
    result[i] = rfr[i].predict(X_test)
```

```python
finalResult = []
for i in range(len(result[0])):
    for j in range(len(result)):
        if (i < len(result[j])):
            finalResult.append(result[j][i])
```

```python
plt.plot(finalResult[-300:])
y = dataset[:,1]
n = len(y)
b_new = -25
b_train = round(n * 0.75)
y_test = y[b_train:b_new]
plt.plot(y_test[-300:])
```
